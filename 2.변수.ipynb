## 알고리즘이란?
- 문제 해결에 필요한 기본적 연산(operation)들을 정확한 순서로 나열하는 것

## 알고리즘의 구현 시 고려사항들
- 기본적 연산: 보통 사칙연산이나 대소 비교 등 + Python에서 제공하는 내장함수

#### "a"가 정말로 "label"인가?

- 1) a = [1,2,3], b=a로 정의
    - 여기서 "object"는 "[1,2,3]" 이라는 배열
- 2) a가 정말로 label이라면, b=a를 통해 b라는 새로운 label을 달아준 것

- 3) a와 b가 가리키는 [1,2,3]이라는 값을 갖는 동일한 "object"일 것

- 4) a가 가리키는 "object"의 값을 바꾸면, b는 동일한 "object"에 대한 다른 이름일 뿐이므로, b가 가리키는 값도 바뀌어야 함

- 이를 아래의 예제로 확인

```python
a = [1, 2, 3]
print('a, before change: ', a)
b = a
print('b, before change: ', b)

a[0] = '_str'
print('a, after change: ', a)
print('b, after change: ', b)

# --> a의 값만 고쳤는데 b의 값까지 바뀐 것을 확인 가능
```

```python
print(hex(id(a)))
print(hex(id(b))) # 메모리 주소가 같음을 확인할 수 있음

# b is just another name of a; contained values are the same
```

- a와 b가 같은지 확인할 때,
    - 1) 같은 object인지 확인 ("is") (같은 메모리 주소를 가짐)
    - 2) 다른 object이지만 object 내부의 값이 같은지 확인 ("==") (다른 메모리 주소를 가짐)

- 의 두 가지 방법이 있음

일반적으로는 2)를 필요로 하지만, 상황에 따라 1)이 필요할 수 있음
- a = [1,2]일 때 b의 초기값을 a로 놓기 위해 "b=a"를 사용하면, 위처럼 a의 값을 고쳤는데 의도치 않게 b의 값이 바뀔 수 있음
- 이를 "is", "=="를 통해 확인가능

```python
a = 1e+1
b = a
print("a==b?: ", a==b)
print(hex(id(a)), hex(id(b)))
print("is a b?: ", a is b)
print('\n')

a = 1e+1
b = 1e+1
print("a==b?: ", a==b)
print(hex(id(a)), hex(id(b)))
print("is a not b?: ", a is not b)
```

주의: [-5, 256]은 Python에서 최적화를 위해 미리 메모리 주소가 할당되어 있음

- "==": value가 같은지 다른지를 확인
- "is": 같은 object인지를 확인


- 다른 object여도 값이 같을 수 있음
- "=" (assignment operator)를 잘 못 사용하면 의도치 않게 다른 변수가 같은 object를 가리키고 있을 수 있음

#### text formatting
- (1) % 혹은 (2) {}와 .format, (3) f-string을 통해 string을 다룰 수 있음


- ##### (1) using %
    - *string* % *data*를 통해 string의 형태로 data를 print할 수 있음
    - 앞서 언급했던 magic function과는 다름
    
| format | description       |
|--------|-------------------|
| %s     | string            |
| %d     | decimal integer   |
| %f     | decimal float     |
| %e     | exponential float |
| %%     | literal %         |


```python
x = 26
y = 42
print('x is: ', x, 'and y is: ', y)
print('x is %s, y is %s' % (x,y))
```

```python
# examples
print('%s' % 26.3)
print('%d' % 26.3)
print('%f' % 26.3)
print('%e' % 26.3)
print('%d%%' % 26.3)
```

- 최소 field width, 최대 character width와 정렬 등을 지정할 수도 있음

```python
n=42; f=3.14; s='long string'
'%6.2d %6.2f %6.2s' % (n, f, s) # 6: 최소 field width, 2: 최대 character width (or decimal points)
```

```python
'%6.2d %6.2f %-10.8s' % (n, f, s) # -: 왼쪽으로 정렬, "10.8": 10글자만큼의 공간을 할당하되, print하는 string의 길이는 8
```

```python
'%*.*d %*.*f %-*.*s' % (6, 2, n, 6, 2, f, 10, 8, s)
```

##### (2) {} and format

```python
print('{} {} {}'.format(n, f, s), '\n')
print('{n} {f} {s}'.format(n=42, f=3.14, s='long string'), '\n')
print('{f} {s} {f}'.format(n=42, f=3.14, s='long string'), '\n')


print('{n:d} {f:f} {s:s}'.format(n=42, f=3.14, s='long string'), '\n')
print('{0:6d} {1:4.2f} {2:12.10s}'.format(n, f, s), '\n')
print('{:6d} {:4.2f} {:12.10s}'.format(n, f, s), '\n')
print('{2:6d} {1:4.2f} {0:12.10s}'.format(s, f, n), '\n')
```

##### (3) using f-string

```python
x = 10
y = 3
print(f'{n} {f} {s}')
print('x+y = %d | x*y = %d' % (x+y, x*y))
print('x+y = {} | x*y = {}'.format(x+y, x*y))
print(f'x+y = {x+y} | x*y = {x*y}')

print(f'{x:.2f}')
```

```python
for x in range(11):
    print(f'{x} {x:02} {x*x:3} {x*x:.2e} {x**3:4} {x**3:4.6f}')
```

```python
s = 'abcd'

print(f'{s}') # left ordering by default
print(f'{s:<10}') # left ordering by default
print(f'{s:>10}') # right ordering with ">"; total len == 10
```
