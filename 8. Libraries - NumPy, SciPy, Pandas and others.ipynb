## Useful functions
- np.zeros() # 영행렬
- np.ones() # 모든 원소가 1인 행렬
- np.identity() # 단위행렬
    - np.eye() # 단위행렬. non-square matrix 및 1의 값을 off-diagonal에 생성 가능


- np.linspace(start, stop, num=50, ...) # [start, stop]의 interval을 num의 개수만큼 쪼갬.


- np.shape() # 행렬의 shape를 반환
    - len(), size()
- np.reshape(a, newshape) # 행렬을 reshape함


- np.concatenate((a1, a2, ...), axis=0) # a1, a2, ...를 concatenate함


- np.where(condition, [x, y]) # condition이 True일 때 x를, False일 때 y를 반환함 (condition은 element-wise로 check)

```python
x1 = np.array([[0,1,2], [2,3,4]])
print(x1)
print('-'*80)

print(x1.shape)
print(x1.size) # number of elements in array
print(len(x1))
print(len(x1.T))

print('-'*80)
print(x1.reshape([3,2]))
```

```python
# np.linspace() returns evenly spaced numbers over a specified interval

print(np.linspace(0, 10, 9))
print(np.linspace(start=0, stop=10, num=9))
print(np.linspace(stop=10, start=0, num=11)) # increasing
print('-'*80)
print(np.linspace(10, 0, 11)) # decreasing
```

```python
# np.concatenate() joins a sequence of arrays along an existing axis
x1 = np.identity(2)
x2 = np.array([[1,2], [2,1]])
print(np.concatenate((x1, x2), axis=0), '\n')
print(np.concatenate((x1, x2), axis=1), '\n')
print(np.concatenate((x1, x2), axis=None), '\n')
```

```python
# identity matrix
print(np.eye(3)) # equivalent to np.identity(3)
print(np.eye(3, 2))
print(np.eye(2, 3))

print(np.eye(3, k=1), '\n') # upper off-diagonal
print(np.eye(3, k=-1), '\n') # lower off-diagonal
```

```python
import numpy as np
x = np.identity(2) # equivalent to np.eye(2)
y = np.array([[2, 3], [1, 2]])
print(type(x))
print(type(y))

print(f'x:{x}\n')
print(f'y:{y}\n')

print(f'x*y:{x*y}\n') # element-wise operation
print(f'x*y:{np.multiply(x,y)}\n') # element-wise operation

print('-'*80)
print(f'x*y:{np.dot(x,y)}\n')
print(f'x*y:{x.dot(y)}\n')

print('-'*80)
print(f'x*y:{np.matmul(x,y)}\n')
```

```python
x = np.identity(2)
y = np.array([[2, 3], [1, 2]])
print(f'type(x):{type(x)}, type(y):{type(y)}')
print('-'*80)
x = np.asmatrix(x)
y = np.asmatrix(y)
print(f'type(x):{type(x)}, type(y):{type(y)}')
print(f'x*y:{x*y}')
```

```python
x = np.array(range(3))
print(x)

# element-wise 
print(x.repeat(3))
print(x.repeat(3, axis=0), '\n') # default axis

x = np.array([[1,2], [3, 1]])
print(x.repeat(2, axis=0), '\n')
print(x.repeat(2, axis=1), '\n')

# array-wise
print(np.tile(x, (1,2)), '\n')
print(np.tile(x, (2,1)), '\n')
```

### NumPy array with string
- string 변수를 element로 갖는 numpy array도 정의 가능
- string "matrix"에 대해서 연산을 취할 일이 없기 때문에, 일반적으로는 사용하지 않음
    - numpy array 대신 list 혹은 list of list를 사용해서 변수를 저장

```python
import numpy as np
x = np.array([['a', 'b'], ['c', 'd']]) # "2 by 2"
print(x, type(x))

y = [['a', 'b'], ['c', 'd']] # "1 by 4"
print(y, type(y))
```

### list vs NumPy array
- list: efficient  general-purpose containers
    - efficient insertion, deletion, appending, concatenation, list comprehension
    - do NOT support "vectorized" operations: elementwise addition and multiplication


- NumPy: multidimensional array of objects all of the same type
    - list와는 달리, every element에 대해 type 정보를 저장하지 않으므로 더 efficient

```python
x = np.array([[1, 'a'], [2, 'b']])
print(x)
print(type(x))
print(x[0], type(x[0]))
print(x[0,0], type(x[0,0]))
print(x[0,1], type(x[0,1]))
```

```python
x = np.array([1, 0])
x_ = np.array(['a', 'b'])
print(x[0], type(x[0]))
print(x_[0], type(x_[0]))

x__ = np.concatenate([x, x_]) # concatenate
print(x__[0][0], type(x__[0][0]))
```

```python
y = [[1, 'a'], [2, 'b']]
print(type(y))
print(y[0], type(y[0]))
print(y[0][0], type(y[0][0]))
print(y[0][1], type(y[0][1]))
```

## Optimization

```python
import matplotlib.pyplot as plt

def myQuad(x):
    return (x-1)**2

x = np.linspace(start=-10, stop=10, num=50)
plt.plot(x, myQuad(x))
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)=(x-1)**2')
plt.show()
```

```python
from scipy.optimize import minimize
x_ = minimize(myQuad, x0=-1)
print(type(x_), '\n')
print(x_)
```

```python
def myQuad2(x):
    return (x-3)*(x-1)*(x+1)*(x+3)

x = np.linspace(start=-5, stop=5, num=50)
plt.plot(x, myQuad2(x))
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)=(x-1)(x-3)')
plt.show()
```

```python
x0 = [-2, -1, 0, 1, 2]
for x0_ in x0:
    x_ = minimize(myQuad2, x0=x0_)
    print(f'x0: {x0_}')
    print(f'number of evaluation:{x_.nfev}')
    print(f'solution: {x_.x}\n')
```

# Sympy
- Symbolic math

```python
import sympy as sym
a = sym.Rational(1, 2)
print(type(a), a)

print(sym.pi * 2)
print(sym.pi.evalf())
print((sym.pi + sym.exp(1)).evalf())

print(sym.oo) # infinity
print(-sym.oo) # -infinity
```

```python
x = sym.Symbol('x')
y = sym.Symbol('y')

print(x + y + x - y + y)

print('-'*80)

print(sym.diff(sym.sin(x), x))
eq = (x + y)**2
print(eq)
print(sym.diff(eq, x))

print('-'*80)
print(sym.integrate(3*x**2+5, x))

print('-'*80)
print(sym.solveset(x**2 - 2*x + 1, x))
print(sym.solveset((x-3)*(x-1)*(x+1)*(x+3), x))

print('-'*80)
sol = sym.solve((x - 2*y + 5, -2*x + y +3), (x, y))
print(sol) # dict
```

## DataFrame construction

```python
import numpy as np
import pandas as pd

x = [10, 20, 30, 40] # tuple, NumPy array, ...
df = pd.DataFrame(x, columns=['numbers'], index=['a', 'b', 'c', 'd'])
print(df)
print('-'*80)

print(df.index)
print(df.columns)
print('-'*80)

print(df.loc['c'])
print(df.loc[['a', 'd']])
print(df.loc[df.index[1:3]])
print('-'*80)
```

## Inserting column and appending row

```python
# add a column
df['floats'] = (1.5, 2.5, 3.5, 4.5)
print(df)
print(df['floats'])
print('-'*80)

df['names'] = pd.DataFrame(['Yves', 'Guido', 'Felix', 'Francesc'], index=['d', 'a', 'b', 'c'])
print(df)
print('-'*80)

# append with index
df = df.append(pd.DataFrame({'numbers': 100, 'floats': 5.75, 'names': 'Henry'}, index=['z', ]))
print(df)
```

## Merge and join

```python
# merge with another DataFrame
df.join(pd.DataFrame([1, 4, 9, 16, 25], index=['a', 'b', 'c', 'd', 'y'], columns=['squares', ]))
print(df)
print('-'*80)

# outer join. (inner, left, right can be used instead)
df = df.join(pd.DataFrame([1, 4, 9, 16, 25], index=['a', 'b', 'c', 'd', 'y'], columns=['squares', ]), how='outer')
print(df)
```

## Basic functions
- sum(), cumsum(), mean(), std(), describe()
- head(), tail()

```python
# basic operations
print(df.numbers**2)
print('-'*80)

print(df['squares'].sum())
print('-'*80)

print(df[['numbers', 'squares']].mean())
print('-'*80)

print(df[['numbers', 'squares']].std())
print('-'*80)

print(df['floats'].apply(lambda x: x * 10))
```

```python
print(df.describe())
print('-'*80)
print(df.head())
print('-'*80)
print(df.tail(n=3))
```

# Matplotlib
- Comprehensive library for creating static, animated, and interactive visualizations in Python
- Matplotlib 갤러리에 일반적으로 많이 쓰이는 plot들의 sample code가 제공됨 (https://matplotlib.org/gallery/index.html)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2, 0.01)
y = 1 + np.sin(2 * np.pi * x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel='x', ylabel='y', title='example: simple plot')
ax.grid()
plt.show()
```

```python
x = np.random.randn(100, 2) # normal random number, 100 by 2
y = np.exp(x[:, 1])
x = x[:, 0]

fig, ax = plt.subplots()
cm = plt.cm.get_cmap(name='hsv') # colormap list: https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
# sc = ax.scatter(x, y, c=x, s=np.abs(y)*100, alpha=0.5) # c: color, s: size
sc = ax.scatter(x, y, c=x, s=y*100, alpha=0.5, cmap=cm) # c: color, s: size
plt.colorbar(sc)
ax.set_xlabel('x')
plt.xticks(rotation=45)
ax.set_ylabel('$e^{y}$', fontsize=25) # TeX syntax with $ $
ax.set_title('example: scatter plot')
ax.grid(True)
fig.tight_layout()
plt.show()
```
